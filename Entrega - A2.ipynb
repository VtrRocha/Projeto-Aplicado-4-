# **Referencial TeĂłrico**

## SĂ©ries Temporais

Uma sĂ©rie temporal Ă© uma sequĂȘncia de observaĂ§Ă”es coletadas ao longo do tempo em intervalos regulares. Sua anĂĄlise baseia-se na identificaĂ§ĂŁo de componentes estruturais: **tendĂȘncia** (movimento de longo prazo), **sazonalidade** (padrĂ”es cĂ­clicos de curto prazo) e **resĂ­duo** (variaĂ§Ă”es aleatĂłrias) (BOX; JENKINS; REINSEL, 2015). A decomposiĂ§ĂŁo da sĂ©rie nessas componentes Ă© um passo essencial antes da modelagem preditiva.

## Modelos ARIMA e SARIMA

O modelo **ARIMA** (*AutoRegressive Integrated Moving Average*) combina processos autorregressivos (AR), diferenciaĂ§ĂŁo para estacionarizaĂ§ĂŁo (I) e mĂ©dias mĂłveis (MA), sendo parametrizado por (p, d, q). Sua extensĂŁo sazonal, o **SARIMA**, incorpora termos sazonais (P, D, Q, s), tornando-o adequado para sĂ©ries com padrĂ”es periĂłdicos. Esses modelos sĂŁo amplamente utilizados em previsĂŁo de demanda no varejo (HYNDMAN; ATHANASOPOULOS, 2021).

## Prophet

Desenvolvido pelo Facebook (atual Meta), o **Prophet** Ă© um modelo aditivo de sĂ©ries temporais especialmente robusto a dados faltantes, outliers e mudanĂ§as de tendĂȘncia. Modela explicitamente feriados e sazonalidades mĂșltiplas (Taylor; Letham, 2018), sendo amplamente adotado em aplicaĂ§Ă”es de negĂłcios por sua facilidade de uso e interpretabilidade.

## Trabalhos Correlacionados

Estudos anteriores aplicaram modelos de sĂ©ries temporais Ă  previsĂŁo de vendas no e-commerce com resultados promissores. Fildes et al. (2019) demonstraram que modelos SARIMA superam mĂ©todos ingĂȘnuos (naive) em horizontes curtos de previsĂŁo para produtos de alta rotatividade. JĂĄ Bandara et al. (2020) compararam ARIMA, Prophet e redes neurais recorrentes (LSTM), concluindo que modelos estatĂ­sticos clĂĄssicos apresentam desempenho competitivo quando o volume de dados histĂłricos Ă© limitado, enquanto modelos de deep learning se destacam com sĂ©ries longas e mĂșltiplas variĂĄveis.

## MĂ©tricas de AvaliaĂ§ĂŁo

As principais mĂ©tricas utilizadas para avaliar modelos de forecasting sĂŁo: **MAE** (*Mean Absolute Error*), que mede o erro absoluto mĂ©dio; **RMSE** (*Root Mean Squared Error*), mais sensĂ­vel a grandes erros; e **MAPE** (*Mean Absolute Percentage Error*), que expressa o erro em termos percentuais, facilitando a comparaĂ§ĂŁo entre sĂ©ries de diferentes escalas (HYNDMAN; ATHANASOPOULOS, 2021).
